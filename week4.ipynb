```python
import os
import sys
import cv2 as cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

# Activity 1 - display 2 images simultaneously

```python
#create grid
f, ax = plt.subplots(2,2)

#import/read images
img1 = mpimg.imread('dog.jfif')
img2 = mpimg.imread('soccer.jpg')

#convert to grayscale / read in grayscale
gray1 = cv2.imread("dog.jfif", cv2.IMREAD_GRAYSCALE)
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

# convert BGR to RGB
# opencv different to matplotlib
# opencv read in BGR, plt read in RGB
gray1 = cv2.cvtColor(gray1, cv2.COLOR_BGR2RGB)
gray2 = cv2.cvtColor(gray2, cv2.COLOR_BGR2RGB)

#assign images
imgplot1 = ax[0,0].imshow(img1)
imgplot2 = ax[1,0].imshow(img2)
imgplot1_1 = ax[0,1].imshow(gray1)
imgplot1_2 = ax[1,1].imshow(gray2)

#show images
plt.show()
```

# Activity 2 - video adjustment

```python
# get video
cap = cv2.VideoCapture('img_pexels.mp4')

fourcc = cv2.VideoWriter_fourcc('m', 'p', '4', 'v')
out = cv2.VideoWriter('smaller_img_pexels.mp4',fourcc, 10, (1280,720))

while True:
    ret, frame = cap.read()
    if ret == True:
        b = cv2.resize(frame,(1280,720),fx=0,fy=0, interpolation = cv2.INTER_CUBIC)
        out.write(b)
    else:
        break
    
cap.release()
out.release()
cv2.destroyAllWindows()

print(os.getcwd())
```

# Activity 3 - create a random noise color and grayscale

```python
f, ax = plt.subplots(1,2)

color_noise=np.random.uniform(0,1,(20,20,3))
noise = (800*np.random.random((20,20))).clip(0,255).astype(np.uint8)

noise = cv2.cvtColor(noise, cv2.COLOR_BGR2RGB)

#show images
ax[0].imshow(color_noise)
ax[1].imshow(noise)
plt.show()
```

# Activity 4 - extract image
Extract the region of interest (flower) from the 'flower.jfif'.

```python
# get photo
img_flower = cv2.imread("flower.jfif")

# select ROI
roi = cv2.selectROI(img_flower)

# cropping
roi_selected=img_flower[int(roi[1]):int(roi[1]+roi[3]), int(roi[0]):int(roi[0]+roi[2])]

# show in new window - opencv
cv2.imshow('ROI of the flower', roi_selected)
cv2.waitKey(0)
cv2.destroyAllWindows()

# show in notebook - matplotlib
roi_selected = cv2.cvtColor(roi_selected, cv2.COLOR_BGR2RGB)
plt.imshow(roi_selected)
```

# Activity 5 - enlarge image
Enlarge the image "dog.jfif" by using different techniques: 
1) Linear interpolation
2) Cubic interpolation and 
3) nearest neighbor interpolation. 

```python
# get photo
img_dog = cv2.imread("dog.jfif")

# resize
img_resize1 = cv2.resize(img_dog, None, fx = 2, fy = 2, interpolation = cv2.INTER_LINEAR)
img_resize2 = cv2.resize(img_dog, None, fx = 2, fy = 2, interpolation = cv2.INTER_CUBIC)
img_resize3 = cv2.resize(img_dog, None, fx = 2, fy = 2, interpolation = cv2. INTER_NEAREST)

# show photo
cv2.imshow('normal', img_dog)
cv2.imshow('Linear interpolation', img_resize1)
cv2.imshow('Cubic interpolation', img_resize2)
cv2.imshow('Nearest neighbor interpolation', img_resize3)

print('original image shape:', img_dog.shape)
print('resized image shape:', img_resize1.shape)
print(img_resize2.shape)
print(img_resize3.shape)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
